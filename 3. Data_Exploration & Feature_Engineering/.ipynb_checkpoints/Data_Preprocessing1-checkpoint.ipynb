### Import Libraies

```python
# import the different libraries 

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
Data = pd.read_csv('C:/Users/pooja/OneDrive/Desktop/We cloud/ML/Project #2/3. Data_Exploration & Feature_Engineering/Data_Preprocessing1.csv',index_col='Sales_Date')
```

```python
Data=Data.drop(columns=['Unnamed: 0','Store_id'])
```

```python
Data['Weekly_Sales'].plot(figsize=(15,3))
```

```python
# Simple moving average
Data['df_rolled_mean'] = Data['Weekly_Sales'].rolling(window = 1).mean()
Data['df_rolled_std'] = Data['Weekly_Sales'].rolling(window = 1).std()
```

```python
Data[['Weekly_Sales','df_rolled_mean']].plot()
```

```python
Data[['Weekly_Sales','df_rolled_std']].plot()
```

```python
#Cummilative moving Average
Data['Weekly_Sales'].expanding().mean().plot(figsize=(15,3))
```

### lets check trend of data

```python
plt.figure(figsize=(15,6))
data1 = sns.lineplot(data=Data, x='Year-Month', y="Weekly_Sales")
plt.xticks(rotation = 60)
plt.show()
```

```python
from statsmodels.tsa.seasonal import seasonal_decompose
multi_plot = seasonal_decompose(Data['Weekly_Sales'], model = 'add', extrapolate_trend='freq',period=2906)

plt.figure(figsize=(20,5))
multi_plot.observed.plot(title = 'weekly sales')

plt.figure(figsize=(20,5))
multi_plot.trend.plot(title = 'trend')

plt.figure(figsize=(20,5))
multi_plot.seasonal.plot(title = 'seasonal')

plt.figure(figsize=(20,5))
multi_plot.resid.plot(title = 'residual');
```

### Dickey-Fuller Test

```python
from statsmodels.tsa.stattools import adfuller
```

```python
dftest = adfuller(Data['Weekly_Sales'],autolag='AIC')
```

```python
dfoutput = pd.Series(dftest[0:4],index['Test Statistic','P-value','#lags used','Number of observation used'])
for key,value in dftest[4].items():
    dfoutput['critical value (%s)' %key]=value

print(dfoutput)
```
