### Need some Feature Enginnering to Do more Accurate analysis

```python
#to do more accurate analysis based on month and year
#need to create 2 more column month and year using the date column
Retail_Analysis_Data['Sales_Month']=pd.DatetimeIndex(Retail_Analysis_Data['Date']).month
Retail_Analysis_Data['Sales_Year']=pd.DatetimeIndex(Retail_Analysis_Data['Date']).year
```

```python
#Date is in object format.
#let's conver date into the datetime format
Retail_Analysis_Data.Date = pd.to_datetime(Retail_Analysis_Data.Date)
```
