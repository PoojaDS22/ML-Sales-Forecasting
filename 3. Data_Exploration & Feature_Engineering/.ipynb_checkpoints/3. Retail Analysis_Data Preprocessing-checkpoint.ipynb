```python
#it will take several minutes to run
from statsmodels.tsa.seasonal import seasonal_decompose
multi_plot = seasonal_decompose(Data['Weekly_Sales'], model = 'add', extrapolate_trend='freq',period=2906)

plt.figure(figsize=(20,5))
multi_plot.observed.plot(title = 'weekly sales')

plt.figure(figsize=(20,5))
multi_plot.trend.plot(title = 'trend')

plt.figure(figsize=(20,5))
multi_plot.seasonal.plot(title = 'seasonal')

plt.figure(figsize=(20,5))
multi_plot.resid.plot(title = 'residual');
```

## Rolling Method

```python
Data['Weekly_Sales'].plot()
```

```python
Data['Data_rolled_std'] = Data['Weekly_Sales'].rolling(window=2948,min_periods=1).std()
Data['Data_rolled_mean'] = Data['Weekly_Sales'].rolling(window=2948,min_periods=1).mean()
#print(Data_rolled_mean,Data_rolled_std)
```

```python
#plot Rolling Statastics
plt.figure(figsize=(18,6))
#orig = plt.plot(Data,color='blue',label='Original')
mean = plt.plot(Data['Data_rolled_mean'],color='red',label='Rolling Mean')
std = plt.plot(Data['Data_rolled_std'],color='black',label='Rolling Std')
#plt.legend(loc="best")
#plt.show(black=False)
plt.xticks(rotation = 90)
plt.show()
```
